```python
# Load env variables # Create an API client

from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = { 
        "role": "user", 
        "content": text 
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = { 
        "role": "assistant", 
        "content": text 
    }
    messages.append(assistant_message)

def chat(messages, stop_sequences=[], system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "stop_sequences": stop_sequences,
    }
    if system:
        params["system"] = system
       
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

system = "Respond only with a numbered list. Start directly with '1. '. No introduction, no explanation."

add_user_message(messages, "Give me three useful Linux commands. Just the command names.")

text = chat(messages, stop_sequences=["\n4."], system=system)
print(text)
```
